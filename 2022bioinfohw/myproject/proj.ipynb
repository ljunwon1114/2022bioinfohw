HDGI@ 등 삭제 

```python
toremove = re.compile('\^.')
pileup['matches'] = pileup['matches'].apply(lambda x: toremove.sub('', x))
pileup
```

```python
pileup = pileup[['chrom', 'pos', 'matches']]
pileup
```

```python
for base in ["a", "t", "g", "c"]:
    pileup[base] = pileup.matches.str.count(base)
pileup['base'] = pileup[["a", "t", "g", "c"]].sum(axis=1)
pileup
```

#Shannon Entropy 계산

```python
pileup[["a", "t", "g", "c"]] = pileup[["a", "t", "g", "c"]].div(pileup['base'], axis = 0)
pileup = pileup.dropna()
pileup
```

```python
import numpy as np
prop = pileup[["a", "t", "g", "c"]].to_numpy()
pileup["entropy"] = -np.sum(prop * np.log(prop+1e-9), axis=1)
print(type(pileup["entropy"]))
pileup.loc[pileup["entropy"] <= 0, "entropy"] = 0
pileup
```

```python
pileup['pos2']=pileup['pos']+1
pileup = pileup[['chrom', 'pos', 'entropy']]
pileup
```
